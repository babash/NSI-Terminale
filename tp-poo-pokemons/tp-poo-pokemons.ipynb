````
import json
import requests
import matplotlib.pyplot as plt
import matplotlib.image as img
from PIL import Image
import io
from IPython.display import Audio, display
````

!!! note Création d'une classe POKEMON
L'objectif de ce TP est de créer un programme en objet dans lequel nous allons créer des Pokémons et les fair interagir entre eux.
Pour commencer, nous créons la classe pokémon ci-dessous. 
!!!

````
class Pokemon:

    def __init__(self, name, pid, height, weight, types, stats, sprite, cry):
        self.name = name
        self.id = pid
        self.height = height
        self.weight = weight
        self.types = tuple(types)
        self.stats = stats
        self.sprite = sprite
        self.cry = cry

    def show(self):
        # Fetch the image data from the URL using requests
        response = requests.get(self.sprite)
        #response.raise_for_status()  # Raise an exception if the request fails

        # Open the image data with Pillow
        image = Image.open(io.BytesIO(response.content))

        # Convert the Pillow image to a NumPy array for matplotlib
        image_np = img.pil_to_array(image)

        plt.imshow(image_np)
        plt.axis('off')
        plt.show()
        if self.cry:
            response = requests.get(self.cry)
            if response.status_code == 200:
                audio = Audio(response.content, autoplay=True)
                display(audio)
            else:
                print("Error retrieving cry audio.")
        else:
            print("No cry available for this Pokémon.")
````

!!! note Création d'un objet POKEMON
Puis nous créons un pokemon spécifique, en affectant correctement les attributs définis dans le constructeur de la classe pokemon. 
!!!

````
pikachu = Pokemon(name = 'pikachu', 
                   pid = 25, 
                   height = 4, 
                   weight = 60,
                   types = ('electric',),
                   stats = {'hp': 35,
                             'attack': 55, 
                             'defense': 40, 
                             'special-attack': 50, 
                             'special-defense': 50, 
                             'speed': 90},
                   sprite = 'https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/25.png',
                   cry = 'https://raw.githubusercontent.com/PokeAPI/cries/main/cries/pokemon/latest/25.ogg')
pikachu.show()
````

````
def pokemon_from_data(pokemon_data):
    return Pokemon(pokemon_data['name'],
                   pokemon_data['id'],
                   pokemon_data['height'],
                   pokemon_data['weight'],
                   tuple(pokemon_data['types']),
                   pokemon_data['stats'],
                   pokemon_data['sprite'],
                   pokemon_data['cry'])
````
